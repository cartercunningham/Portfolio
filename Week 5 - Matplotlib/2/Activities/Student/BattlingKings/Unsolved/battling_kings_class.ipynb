```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read CSV
got_data = pd.read_csv("Resources/got.csv")
got_data
```

```python
# Print column names & data types 
print(got_data.columns) 
got_data.dtypes

# Get attacker and defender data
attackers = got_data["attacker_king"].value_counts()
defenders = got_data["defender_king"].value_counts()
print(attackers)
print(defenders)
```

```python
# Get total battle data
battle_fields = attackers.add(defenders, fill_value=0)
```

```python
# Configure plot and ticks
battle_fields.plot(kind="bar")

# Set textual properties
plt.title("The BLoodthrist of Kinds")
plt.xlabel("number of Battles Participated In")
plt.ylabel("King")
plt.show()
plt.tight_layout()
```
