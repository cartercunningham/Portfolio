# Miles Per Gallon

There are many different features of a car that can help determine how many miles per gallon it has. In this activity, we will be creating a scatterplot to chart some of these relationships.

* The scatterplot that we want to create will compare 'mpg' to 'horsepower'. A reference image can be found at the bottom of this notebook, but how we go about creating this chart is completely up to you!

    * When reading in the data, there are 6 rows that are missing values in 'horsepower'. It is up to you to figure out what the author put in their place and drop the rows.

```python
# Check out books from Library 
import matplotlib.pyplot as plt
import csv 
import pandas as pd 
import numpy as np
```

```python
# Read CSV 
cars = pd.read_csv("../Resources/mpg.csv")
print(cars.count())
print(cars.columns)
print(cars.dtypes)
```

```python
# Remove missing values from horsepower 
print(cars.count())
cars = cars.loc[cars['horsepower'] != '?']
cars.head()
```

```python
# Set Car as the index 
cars = cars.set_index('car name')
cars.head()
```

```python
# Remove origin column 
del cars['origin']
print(cars.dtypes)
# Convert horsepower to numeric 
cars['horsepower'] = pd.to_numeric(cars['horsepower'])
print(cars.dtypes)

# Create Scatter plot 
cars.plot(kind="scatter", x="horsepower", y="mpg", grid=True, figsize=(20,10), title="Horsepower Vs. MPG")
```
