```python

# Dependencies# Depen 
import numpy as np
import matplotlib.pyplot as plt
```

```python
# set X axis to numerical values for the month 
x_axis = np.arange(1, 13)
x_axis
```

```python
# Average weather temp# Avera 
y_axis = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
```

```python
plt.plot(x_axis, y_axis)
```

```python
# Convert to Celsius C = (F-32 * 0.56)
celsius = [(temp-32)* 0.56 for temp in y_axis ]
celsius
```

```python
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Plot both on the same chart 
plt.plot(x_axis, y_axis)
plt.plot(x_axis, celsius)
```
